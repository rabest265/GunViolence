```python
import numpy as np
import pandas as pd
from os import path
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt

from PIL import Image
import urllib
import requests
```

```python
words="Guns VIOLENCE innocent targeted WorkPlace MassShootings Random shelter Mass college DriveBy strayBullet crossfire Fire gunfire gunshot pop MentalHealth emergency police DomesticViolence rifles Ammo School VICTEMS shooting accidental suicide Gangs GangViolence murder SelfInflicted FEAR bullets Wounded killed ArmedRobbery HATE CRIME terrorism "

mask = np.array(Image.open(requests.get('http://clipart-library.com/img1/1588392.jpg', stream=True).raw))
```

```python
wc = WordCloud(background_color="white", mask=mask,
               stopwords=STOPWORDS,colormap="Greys",  contour_color="grey", contour_width=1, max_font_size=125, min_font_size=20 ).generate(words)



# store to file
wc.to_file(path.join("GunCloud.png"))
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
```
